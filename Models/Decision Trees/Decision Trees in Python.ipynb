### CART
**C**lassification **a**nd **R**egression **T**rees:
* Decision trees can perform classification and regression solutions
* They work by splitting your data into more pure sub-groups at each stage
* Purity is measured with information gain algorithms (e.g. Gini score, entropy)
* Trees are made up of nodes, beginning with the root node where you have all your data (and all classes), moving through decision nodes where data is split into distinct groups until finally reaching leaf nodes
* At each node, a feature is selected and an attribute threshold is used to split the data
    * The feature selected is the one which gives greatest information gain (i.e. best purity at subsequent leaf nodes)
    * e.g. split Iris dataset based on petal length being <= 4.5cm
* Purity is defined as the mix of classes in the next set of nodes (e.g. if your parent node has 2 classes split 50/50 and each this is perfectly split so that your two child nodes have 1 class only then this is the maximum purity possible)
* For classification problems, the output class variable is used to measure purity/information gain whilst in regression tasks continuous numerical outputs are assessed instead
* Each node acts in a greedy manner to select the best feature at that moment in time which will yield the largest information gain for categorical targets
* If no maximum depth is set, decision trees will continue splitting the dataset until perfect purity is achieved, a minimum sample size is hit or you run out of data to split. This leads to overtraining (one of the drawbacks of decision trees) and as such you must set a maximum depth for your tree (called pre-pruning) to avoid this
* Other pruning steps exist to optimize your decision tree model, there are many but some exampes include recursively/retrospectively adjusting rule thresholds, node order, features selected at each node etc. to maximize information gain at each stage and overall

Decision Tree Algorithms and SKLearn:
* There are several different decision tree algorithms which implement different methods (e.g. pruning steps, order, feature selection rules etc.)
* SKLearn implements an optimized CART method, similar to the C4.5 method (detailed notes in below link)
* One current limitation of this method is that it doesn't support categorical variables at present

### Loading the Data (Iris)

```python
# load libraries
import pandas as pd
from sklearn.datasets import load_iris

# load the dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# check data
df.head()
```

```python
# load libraries
from sklearn.model_selection import train_test_split

# split train and test data
X_train, X_test, y_train, y_test = train_test_split(df[data.feature_names], df['target'], random_state = 0)
```

### Building the Model
Model Notes:
* Decision trees aren't sensitive to non-normalized data (unlike PCA, regression etc.) therefore we can skip this pre-processing step
* We can use the Gini score or entropy to determine which features to select and calculate the information gain at each step
    * Both measure the purity of class/separation/information gain at each stage
    * They implement different formulas/algorithms to achieve this
    * However, tests show that they agree to a high accuracy almost all the time, so you can pretty much pick either and get the same results
    * Gini tends to be more popular because it's quicker (entropy uses logs in its formula, slowing it down)
    * Therefore, we will use Gini here
* You can choose the 'splitter' method also to tune your model
    * 'best' selects the optimal features and order from all combinations to optimize information gain
    * 'random' selects random combinations which can help to reduce overfitting and also requires less processing as it doesn't run all combinations of features, orders etc.
* Max tree depth is key in over/underfitting also. If you set no limit, your tree will expand until all possible leaves are created, resulting in a very complex model which is overfitted to your train data and generalizes poorly in the real world. Setting a depth too low though can result in underfitting, where you don't capture enough information about your dataset. A good way to pick an optimal depth is to let your model auto-select a max depth and then experiment slightly higher and lower, analyzing the score of your model and the split of your data that it produces
* [Decision Tree Hyperparameter Tuning](https://towardsdatascience.com/how-to-tune-a-decision-tree-f03721801680)

```python
# load libraries
from sklearn.tree import DecisionTreeClassifier

# build model (set initial depth, we will tune later)
clf = DecisionTreeClassifier(max_depth=2, random_state=0)

# fit model to train data
clf.fit(X_train, y_train)

# predict test labels
#clf.predict(X_test)

# assess score of model
score = clf.score(X_test, y_test)
print(score)
```

### Tuning the Tree
The intial run with no tuning (except an initial depth of 2) gave us a score of 89% which is pretty high.

We will now experiment with parameter tuning to see if we can improve this score.

```python
# experiment with depth ranges
depths = list(range(1, 6))

# iterate through depths
for depth in depths:
    # build model (set initial depth, we will tune later)
    clf = DecisionTreeClassifier(max_depth=depth, random_state=0)

    # fit model to train data
    clf.fit(X_train, y_train)

    # predict test labels
    #clf.predict(X_test)

    # assess score of model
    score = clf.score(X_test, y_test)
    print(f"Score with max_depth = {depth} is {round(score, 2)}")
```

Notes:
* We can see that beyond a depth of 3 the accuracy doesn't increase, so we'll select a depth of 3 as this is the least complex (and therefore least overfitted/sensitive) model with the highest accuracy
* Note that the depth of the decision tree itself doesn't change with the max_depth parameter, we are simply choosing a point to cut off the tree to reduce unnecessary complexity
    * i.e. the tree will always run to it's natural conclusion (e.g. a depth of 4) if allowed, so providing a max_depth of 5 produces an identical model to one with a max_depth of 4 because the tree doesn't need to create 5 levels of complexity to solve the problem
* As well as max_depth, you can also set values for min_samples_leaf or max_leaf_nodes for example to set other parameters which will tune/prune your model (experimenting with different values of these parameters using GridSearch, random search, Bayesian optimization etc. will let you pick an optimal model)

### Feature Importance
Feature Importance:
* Decision trees are quite easy to interpret, feature importance is one reason for this
* You can score the importance of each feature in your model to see which one(s) influenced model decisions the most
* Feature importance values are normalized so that the sum of all importances comes to 1
* Below, we can see the feature importance of our dataset

```python
# load libraries
import numpy as np

# calculate feature importance
importances = pd.DataFrame({'feature':X_train.columns, 'importance':np.round(clf.feature_importances_, 3)})

# show importances
importances = importances.sort_values('importance', ascending=False)
importances
```

Notes:
* You can see here that petal width is the most important with petal length behind it and sepal width with a minor influence on the overall model
* It's important to note that these importances don't necessarily mean that e.g. sepal length has no influence on the class decision in your data, it's simply that the model we created chose petal width first and assigned a high importance to it
* For example, each feature might be important in  determining class, but the model only needed 2 nodes/depths to accurately classify our data, therefore the remaining features were set to low importance to our model, even if a good correlation is present
* Also, feature importance may be lower if 2 features are identical to each other or highly correlated
* Plus feature importance tells us nothing about which class a feature is able to predict well or what the relationships between features are
* As such, averaging feature importances across multiple decision trees (i.e. train/test splits) is a more robust method of determining actual feature importance as well as combining with other analysis (e.g. feature correlation) for a more reliable overview

### Visualizing our Tree
Before concluding this analysis, let's visualize our decision tree to give us the full picture of exactly what's going on and what features, split points etc. have been determined by our model.

[Visualizing Decision Trees](https://towardsdatascience.com/visualizing-decision-trees-with-python-scikit-learn-graphviz-matplotlib-1c50b4aa68dc)

```python
# load libraries
from sklearn import tree
import matplotlib.pyplot as plt

# plot decision tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (3,3), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

### Conclusions
Pros:
* No requirement for feature scaling/normalization/standardization
* Easy to interpret (visualization, feature importances etc.)
* Many options for tuning/optimization
* Handle high dimensional data very well
* Good for classification, regression, predicting missing values, feature selection etc.
* Relatively fast and efficient
* Easily captures non-linear patterns
* Non-parametric and white box (i.e. easy to see inner workings, thresholds etc.)

Cons:
* Prone to overfitting if not properly tuned (e.g. set max_depth)
* Sensitive to noisy data (due to overfitting mentioned above)
* Biased with imbalanced datasets, therefore class balancing required
* Bagging and boosting required to prevent sensitivity to new/slightly different input data

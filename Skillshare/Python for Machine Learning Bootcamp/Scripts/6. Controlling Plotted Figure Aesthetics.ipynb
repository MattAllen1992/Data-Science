```python
# load libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# plot in-line
%matplotlib inline
```

```python
# define sinplot
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100) # start, stop, # of steps
    for i in range(1, 7):
        plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip) # not really sure...

# plot sinplot
sinplot()
```

```python
# adjust style (shift + tab for options)
sns.set_style('white', {'axes.grid':True, 'xtick.direction':'in'})
sinplot()
sns.despine(left = True, bottom = False) # remove/set axis lines
```

```python
# check current options
sns.axes_style()
```

```python
# change style
sns.set_style('darkgrid')

# tweak pre-configured settings
sns.set_context('talk', font_scale=1.5)
sinplot()
```

```python
# check pallette
current_palettes = sns.color_palette()
sns.palplot(current_palettes)
```

```python
# crop colour palette and tweak colours etc.
sns.palplot(sns.color_palette('hls', 8))
```

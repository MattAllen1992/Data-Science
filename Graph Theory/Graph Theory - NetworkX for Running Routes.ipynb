### 1) Brief

Notes:
* The Giantmaster data shows all running routes across a region of mountains
* The **edges are the running routes** themsleves and the **nodes are the start/finish locations** (or crossover points) of each route
* Each edge has a weight which is the distance of the route itself (i.e. **edge-weighted graph**)
* This data is **undirected** (i.e. A >> B == B >> A) so that route directions make no difference
* Nodes are said to be **even-degree nodes** if they have an even number of edges feeding into them, and **odd-degree** if not
* The ambition is to find a **Eulerian solution/tour** so that we can identify a solution where we **travel along each edge once and only once** whilst visiting every node in the system. The fall back would be to identify the shortest possible route between all nodes (where we travel along each edge ideally once, but maybe more if necessary)

### 6) Graph Algorithms
* Now that we have our graph plotted and all data stored
* We can perform certain algorithms to tell us things like:
    * Shortest path between certain nodes
    * Distance between specific nodes
    * Average distance between nodes
    * Fastest route from A to B (given A and B)
    * Find nodes with odd or even degrees
    * Add/remove edges to enable a Eulerian tour (i.e. no duplicate journeys)
    
**Finding Odd-Degree Nodes**
* This can be useful when trying to solve a Eulerian route
* Over half of the nodes in this graph have an odd number of degrees
* This is mostly dead-end routes (i.e. 1 degree)
* But there are also some with 3 or 5 degrees where multiple routes cross through them

```python
# extract nodes with odd degrees
nodes_odd_degree = [node for (node, val) in g.degree() if val % 2 == 1]

# check output
nodes_odd_degree[:3]
```

```python
# nodes with odd degrees
print(f'{len(nodes_odd_degree)} nodes with odd number of degrees.')

# all nodes
print(f'{len(g.nodes)} total nodes in graph.')
```

**Why Calculate Odd-Degree Node Pairs?**
* In order to perform a Euler trail (i.e. visit each node once, achieve the shortest possible distance in doing so)
* We need to have an even number of degrees across all nodes
* As such, we isolate our odd-degree node pairs so that we can fix them
* To do this, we simply calculate a complete graph of all of our odd-degree nodes
* Then we augment our original graph with a copy of these odd-degree node pair paths so that for each odd-degree node, we have doubled the edges between them, resulting in an even-degree for all nodes
* The only downside to this is that we now have parallel/multiple edges between some nodes (resulting in a MultiGraph)

**Minimum Distance Pairs**
* The idea here is to identify all odd node pairs
* Then we can identify the distance between each pair in order to compute the shortest path between them (using Dijkstra's algorithm)
* We then create a complete graph where all nodes are connected to every other node by a unique edge (i.e. ensure 1 route between each node for Euler tour)
* Then we calculate the minimum weight matching (i.e. find shortest routes between nodes, aiming to visit each node only once)
* Note that Networkx only implements max weight matching, therefore we flip our distances to create weights (multiply by -1 so that large numbers represent short distances and small numbers represent large distances) so that when Networkx finds the 'max' weight, it's actually finding the minimum distance
* [Matching Explanation](https://brilliant.org/wiki/matching/)

```python
# find odd node pairs (graph is undirectional so only need to capture once)
odd_node_pairs = list(itertools.combinations(nodes_odd_degree, 2))

# check pairs
odd_node_pairs[:3]
```

```python
# function to find shortest path between pairs
def get_shortest_path_distance(graph, pairs, edge_weight_name):
    # create empty dict
    distances = {}
    
    # iterate through pairs
    for pair in pairs:
        # store {(node1, node2):distance} 
        distances[pair] = nx.dijkstra_path_length(graph, pair[0], pair[1], weight=edge_weight_name)
        
    # return result
    return distances

# compute shortest paths
odd_node_pairs_shortest_path = get_shortest_path_distance(g, odd_node_pairs, 'distance')

# show results
dict(list(odd_node_pairs_shortest_path.items())[:3])
```

```python
# function to create a complete graph with inverted weights
# pair_weights is the output of the above function (i.e. a dictionary of node tuples and distance)
def create_complete_graph(pair_weights, flip_weights=True):
    # create empty graph
    g = nx.Graph()
    
    # iterate through node pairs and distance between them
    for nodes, weight in pair_weights.items():
        # flip distance if specified (to produce weights)
        wt_i = -weight if flip_weights else weight
        
        # add edges to graph, keeping distance as well as flipped weight
        g.add_edge(nodes[0], nodes[1], attr_dict={'distance': weight, 'weight': wt_i})
        
    # return results
    return g

# generate complete graph
g_odd_complete = create_complete_graph(odd_node_pairs_shortest_path, True)

# check output
print(f'Complete graph nodes: {len(g_odd_complete.nodes)}')
print(f'Complete graph edges: {len(g_odd_complete.edges)}')
```

Notes:
* Below we have the same nodes (by location) as in our original graph
* However, the edges between them are now not real routes (i.e. not actual running routes)
* Instead, they represent a complete graph where each node is connected to each other node by exactly one degree/edge
* This essentially allows us to plot all possible routes between our odd-degree edges
* We will now compute our minimum weight matching (i.e. of all the below complete graph edges, what are the minimum routes between each when matching - i.e. when visiting each node exactly once)
* i.e. of the 41 nodes below, what are the 20 or 21 (i.e. 41/2) optimal edges to connect each node with the minimum distance
* [Advanced Matching Algorithms](https://networkx.org/documentation/networkx-1.10/_modules/networkx/algorithms/matching.html#max_weight_matching)

```python
# plot complete graph
plt.figure(figsize=(8, 6))
pos_random = nx.random_layout(g_odd_complete)
nx.draw_networkx_nodes(g_odd_complete, node_positions, node_size=20, node_color='red')
nx.draw_networkx_edges(g_odd_complete, node_positions, alpha=0.1)
plt.axis('off')
plt.title('Complete Graph of Odd-Degree Nodes')
plt.show()
```

```python
# compute minimum weight matching
# note that Networkx max weight matching uses the 'weight' attribute
# by default so we don't need to specify that explicitly here
odd_matching = nx.algorithms.max_weight_matching(g_odd_complete, True)

# show results
print(f'Number of edges in matching: {len(odd_matching)}')

# peek at data
odd_matching
```

```python
# plot minimum weight matching graph
plt.figure(figsize=(8, 6))

# plot complete graph of odd-degree nodes
nx.draw(g_odd_complete, pos=node_positions, node_size=20, alpha=0.05)

# overlay min weight matching onto complete graph
g_odd_complete_min_edges = nx.Graph(odd_matching)
nx.draw(g_odd_complete_min_edges, pos=node_positions, node_size=20, edge_color='blue', node_color='red')

# finalise and show
plt.title('Min Weight Matching on Complete Graph')
plt.show()
```

### 7) Graph Augmentation
Notes:
* Now that we know our shortest paths between odd-degree nodes
* We can add this information to our original graph
* We will do so using a multigraph (a graph where multiple parallel edges can connect the same 2 nodes)
* We will also tag/flag the augmented edges from our minimum distance matching for use later on (i.e. we need to remember which edges these are)

```python
# function to augment original graph with odd-degree node matching edges
# graph is the original input graph
# min_weight_pairs are the odd_matching pairs from the previous step
def add_augmenting_path_to_graph(graph, min_weight_pairs):
    # create multigraph copy of graph to allow parallel edges
    graph_aug = nx.MultiGraph(graph.copy()) # copy to avoid in-place changes
    
    # iterate through matching node pairs
    for pair in min_weight_pairs:
        # add edge to graph with distance and label
        graph_aug.add_edge(pair[0], pair[1],
                           attr_dict={'distance': nx.dijkstra_path_length(graph, pair[0], pair[1]),
                                      'trail': 'augmented'})
        
    # return result
    return graph_aug
        
# augment graph with odd_matching pairs
g_aug = add_augmenting_path_to_graph(g, odd_matching)

# show new node and edge numbers
print(f'Number of edges in original graph: {len(g.edges)}')
print(f'Number of edges in augmented graph : {len(g_aug.edges)}')
```

**Check All Nodes Are Even-Degree**
* Now that we have isolated our odd-degree nodes
* And augmented our original graph (i.e. added on top) with the shortest possible paths of our odd-degree nodes
* Any originally odd-degree nodes will now have even-degrees due to the addition of parallel paths
* The below code shows that the number of degrees of our graph nodes is either 2, 4 or 6 (i.e. all even)
* This allows us now to calculate a Euler trail (which requires all nodes to be even-degree)

```python
# create empty dict
degree_count = {}

# count degrees
for node, degree in g_aug.degree():
    if degree in degree_count.keys():
        degree_count[degree] += 1
    else:
        degree_count[degree] = 1
        
# show results
degree_count
```

### 8) Eulerian Circuit
* The Euler path requires that each edge is traversed exactly once and that this is possible if all nodes are even-degree
* Networkx provides the eulerian_circuit method to perform this
* However, there are a few limitations to consider:
    * Our duplicated/augmented paths don't actually exist in real life/the original graph, so we must breakdown the augmented paths into real alternatives from the original graph (e.g. aug_C = real_A + real_B)
    * When multiple edges exist between nodes (due to augmentation), we must keep a record of which ones have been traversed already (in the node attributes) to prevent using the same one twice. Networkx eulerian_circuit only tells you the order of nodes traversed, it doesn't store a flag of whether or not we've traversed them already, so we'll need to fix this ourselves
    * In theory, you can actually have 2 odd-degree nodes in your graph if they are the start and end node. This is because you will then still have reached each node and traversed each edge exactly once, you just won't end up at the start point. We won't worry about that here but it's worth considering

```python
# begin with a simple circuit (i.e. without above limitations handled)
naive_euler_circuit = list(nx.eulerian_circuit(g_aug, source='b_end_east'))

# check that length of our circuit matches length of graph
# i.e. each edge is traversed only once
print(f'Length of circuit: {len(naive_euler_circuit)}')
print(f'Length of original graph: {len(g_aug.edges)}')

# check output
# contains each start/end node as tuple pairs in order
naive_euler_circuit[:3]
```

**Solving the Limitations**
* To overcome the issue of having augmented paths in our naive eulerian route
* We will iterate through our augmented graph and replace augmented paths (i.e. ones which don't exist in the original)
* And then replace these augmented paths with real paths

```python
# function to replace augmented paths with real ones
def create_eulerian_circuit(graph_original, graph_augmented, starting_node=None):
    # create empty path list
    euler_circuit = []
    
    # create naive circuit
    naive_circuit = list(nx.eulerian_circuit(graph_augmented, source=starting_node))
    
    # iterate through naive edges
    for edge in naive_circuit:
        # pass start and end nodes to get edge # and attributes
        edge_data = graph_augmented.get_edge_data(edge[0], edge[1])
        
        # if it's not an augmented route (i.e. exists in original)
        if edge_data[0]['attr_dict']['trail'] != 'augmented':
            # grab edge attributes (using start and end node to index)
            edge_att = graph_original[edge[0]][edge[1]]
            
            # add attributes and edge start/end node to euler circuit
            euler_circuit.append((edge[0], edge[1], edge_att))
            
        # if it is an augmented route (i.e. invented to solve odd-degree nodes issue)
        else:
            # calculate shortest real path from original graph
            aug_path = nx.shortest_path(graph_original, edge[0], edge[1], weight='distance')
            
            # create a list of the start/end node pairs per edge in real path
            # zips one list to an inverted version of itself to create all possible pairs
            aug_path_pairs = list(zip(aug_path[:-1], aug_path[1:]))
            
            # track process for record
            print(f'Filling in edges for augmented path: {edge}')
            print(f"Augmenting path: {' => '.join(aug_path)}")
            print(f'Augmenting path pairs: {aug_path_pairs}\n')
            
            # iterate through real pairs within augmented path
            for edge_aug in aug_path_pairs:
                # get edge attributes
                edge_aug_att = graph_original[edge_aug[0]][edge_aug[1]]
                
                # add attributes to euler circuit
                euler_circuit.append((edge_aug[0], edge_aug[1], edge_aug_att))
                
    # return completed/real euler_circuit
    return euler_circuit
        
# create eulerian circuit
euler_circuit = create_eulerian_circuit(g, g_aug, 'b_end_east')
```

Notes:
* You can see that the length of our Euler circuit is longer than the original/augmented graph
* This is simply because we have replaced the non-real/augmented paths with their real alternatives
* And in most cases (if not all) the real alternatives involve multiple steps

```python
# check length of euler_circuit
print(f'Length of Euler Circuit: {len(euler_circuit)}')
```

Notes:
* Below we can see the euler circuit that the algorithm has picked
* Note that it pays no real attention to the trail type (i.e. skips from one colour to the next)
* We could add in a requirement that our algorithm pays attention to a specific parameter (e.g. trail colour) if we wanted to

```python
# print out entire solution (step by step)
# show first 5 steps only for brevity
for i, edge in enumerate(euler_circuit[:5]):
    print(i, edge)
```

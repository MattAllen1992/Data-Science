# POS Tagging
## 1. Parts of Speech Tagging
**NOTE:** find full tables of coarse and fine-grained POS tagging options at the bottom of this notebook
### POS Tags
* Parts of speech are divided into 2 chunks:
    * Coarse-grained POS (e.g. nouns, verbs, adjectives)
    * Fine-grained POS (e.g. superlative adjectives)
* They are used to tag additional information to your text to help interpret what it means
* English is a complex language:
    * Slight adjustments to the order, frequency or context of words can result in an entirely different meaning
    * Many words occur very rarely
    * Different words can be used to mean similar things
* As such, POS tagging enables us to get a firmer grasp on what the intended meaning of each word, phrase, sentence and so on is
* Below, you can see that Spacy is clever enough to know that 'jumped' is a past tense verb for example

```python
# load libraries
import spacy

# load english language library
nlp = spacy.load('en_core_web_sm')

# create text
doc = nlp(u'The quick brown fox jumped over the lazy dog\'s back.')

# show attributes of tokens
for token in doc:
    print(f'{token.text:{10}} {token.pos_:{10}} {token.tag_:{10}} {spacy.explain(token.tag_)}')

#doc[4].text # pure text
#doc[4].pos_ # coarse-grained POS tag
#doc[4].tag_ # fine-grained POS tag
#spacy.explain(doc[4].tag_) # explanation of POS tag
```

### Tenses and Variations
* Spacy is able to discern between the tense of words as well as other components
* Below is an example of how it interprets present and past tense implementations of 'read'
* The fine-grained POS tags really dig into the specific context and meaning of individual words based on their context and usage

```python
# create text
doc = nlp(u'I am reading books on NLP.')

# examine 'read' specifically
word = doc[2]

# analyse word
print(f'{word.text:{10}} {word.pos_:{10}} {word.tag_:{10}} {spacy.explain(word.tag_)}')
```

```python
# create text
doc = nlp(u'I read a book on NLP.')

# examine 'read' specifically
word = doc[1]

# analyse word
print(f'{word.text:{10}} {word.pos_:{10}} {word.tag_:{10}} {spacy.explain(word.tag_)}')
```

### Tag Counts
* We can extract the count of the different attributes we have
* Below is the code for POS counts (coarse-grained tags i.e. ADJ, NOUN, VERB)
* Below that is the same code but for TAG counts (i.e. fine-grained tags)
* And below that again are the dependency counts
* It's a useful way to extract information from the tagging process implemented by Spacy
* [Spacy POS Tagging Overview](https://spacy.io/usage/linguistic-features)

```python
# create text
doc = nlp(u'The quick brown fox jumped over the lazy dog\'s back.')

# check POS counts (essentially frequency of each POS type)
POS_counts = doc.count_by(spacy.attrs.POS)
print(POS_counts)

# access specific elements from above POS_counts from document vocab
print(doc.vocab[90].text)

# you can do the reverse and check the ID of the token itself
print(doc[0].pos)

# show code, POS tag and count/frequency
for k, v in sorted(POS_counts.items()):
    print(f'{k}. {doc.vocab[k].text:{5}} {v}')
```

```python
# extract tag counts
TAG_counts = doc.count_by(spacy.attrs.TAG)

# show code, POS tag and count/frequency
for k, v in sorted(TAG_counts.items()):
    print(f'{k:{20}} {doc.vocab[k].text:{5}} {v}')
```

```python
# extract tag counts
DEP_counts = doc.count_by(spacy.attrs.DEP)

# show code, POS tag and count/frequency
for k, v in sorted(DEP_counts.items()):
    print(f'{k:{20}} {doc.vocab[k].text:{5}} {v}')
```

## 2. Visualizing POS
### Displacy
* Displacy can be used to generate various visuals to describe the tagging process
* You can customize the appearance of each visual as well as choose which attribute are actually displayed (via the 'style' setting)

```python
# load libraries
from spacy import displacy

# create text
doc = nlp(u'The quick brown fox jumped over the lazy dog\'s back.')

# define options for displacy visuals
# compact shrinks curved connectors to squares/rectangles
options = {'distance':90, 'compact':True, 'color':'yellow', 'bg':'#09a3d5', 'font':'Times'}

# display POS tags and dependencies in text
displacy.render(doc, style='dep', jupyter=True, options=options)
```

```python
# create text
doc = nlp(u'This is a sentence. This is another sentence. This is a sentence which is longer than the previous 2.')

# create list of spans (sentences here)
spans = list(doc.sents)

# visualize in server
# passing a list of spans or docs to displacy lets
# you visualize one visual per sentence/doc
displacy.render(spans, style='dep', options=options)
```
